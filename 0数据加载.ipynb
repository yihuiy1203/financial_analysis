```python
ai = ['人工智能', '商业智能', '机器学习', '深度学习', '图像理解', '投资决策辅助系统', '智能数据分析', '智能机器人', '语义搜索', '生物识别技术', '人脸识别', '语音识别', '身份验证', '自动驾驶', '无人驾驶', '自然语言处理']
bigdata= ['大数据', '数据挖掘', '数据中心', '文本挖掘', '数据可视化', '异构数据', '征信', '增强现实', '混合现实', '虚拟现实']
```

```python
# Function to count the occurrences of keywords in the text
def count_keywords(text, keywords):
    return sum(text.count(keyword) for keyword in keywords)

# Add two new columns to store the total count of AI and BigData keywords
tb['AI_Count'] = tb['ManaDiscAnal'].apply(lambda x: count_keywords(x, ai))
tb['BigData_Count'] = tb['ManaDiscAnal'].apply(lambda x: count_keywords(x, bigdata))
```

```python
tb.sample(6)
```

```python
tb.sort_values(by=['AI_Count'],inplace=True, ascending=False)
tb.reset_index(drop=True, inplace=True)
tb.head(10)
```
